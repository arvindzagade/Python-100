# 37
Question: Create a function that takes a text file as input and returns the number of words contained in the text file. Please take into consideration that some words can be separated by a comma with no space. For example "Hi,it's me." would need to be counted as three words. For your convenience, you can use the text file in the attachment.

```python
def word_count():
    with open('words2.txt','r') as file:
        text= file.read()
        text = text.replace(',',' ')
        string_split = text.split('')
        return len(string_split)
word_count()
```

```python
def count_word():
    with open('words1.txt','r') as file:
        text = file.read()
        text = text.replace(',','')
        string_split = text.split(' ')
        return len(string_split)
count_word()
```

```python
def count_words(filepath):
    with open(filepath, 'r') as file:
        text = file.read()
    text = text.replace(",", " ")
    string_list = text.split(" ")
    return len(string_list)

print(count_words("words2.txt"))
```

```python
import re

def count_words_re(filepath):
    with open(filepath, 'r') as file:
        text = file.read()
    string_list = re.split(",| ", text)
    return len(string_list)

print(count_words_re("words2.txt"))
```

# 41
Question: Create a script that generates a text file with all letters of English alphabet inside it, one letter per line.

```python
import string

with open("letters.txt", "w") as file:
    for letter in string.ascii_lowercase:
        file.write(letter + "\n")
```
